```python
import numpy as np
import matplotlib.pyplot as plt

N=4 
arr = np.random.random((N,N))
plt.imshow(arr)
plt.xticks(ticks=range(N), labels=[f"Individual {i}" for i in range(N)], rotation=15)
plt.yticks(ticks=range(N), labels=[f"Feature {i}" for i in range(N)])
# plt.colorbar()
# plt.show()
for i in range(len(arr)):
    for j in range(len(arr[0])):
        text = plt.text(j, i, round(arr[i, j]*100),
                       ha="center", va="center", color="black")
plt.savefig("./mat2d_cross-sectional.png")
```

```python
arr = np.random.random((N,N))
plt.imshow(arr)
plt.xticks(ticks=range(N), labels=[f"Time {i}" for i in range(N)], rotation=15)
plt.yticks(ticks=range(N), labels=[f"Feature {i}" for i in range(N)])
# plt.colorbar()
# plt.show()
for i in range(len(arr)):
    for j in range(len(arr[0])):
        text = plt.text(j, i, round(arr[i, j]*100),
                       ha="center", va="center", color="black")
plt.savefig("./mat2d_time-series.png")
```
